```python
from collections import Counter
from pandas import Series, DataFrame
import pandas as pd
import pickle
```

```python
with open('c:/data/tenet_x_train.pkl', 'rb') as file:
    tenet_x_train = pickle.load(file)
# 바이너리파일 불러오기
with open('c:/data/tenet_x_test.pkl', 'rb') as file:
    tenet_x_test = pickle.load(file)
# 바이너리파일 불러오기
with open('c:/data/tenet_y_train.pkl', 'rb') as file:
    tenet_y_train = pickle.load(file)
# 바이너리파일 불러오기
with open('c:/data/tenet_y_test.pkl', 'rb') as file:
    tenet_y_test = pickle.load(file)
```

```python
tenet_train = DataFrame({'x': tenet_x_train, 'y': tenet_y_train})
tenet_test  = DataFrame({'x': tenet_x_test, 'y': tenet_y_test})
```

```python
tenet_train = tenet_train[tenet_train.y != '중립']
tenet_test  = tenet_test[tenet_test.y != '중립']
```

```python
#---------------------------------------------------
# 분석 시작

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
cv2 = CountVectorizer(ngram_range=(1,1))
x_train_cv = cv2.fit_transform(tenet_train.x)    # 학습시킴
```

```python
with open('c:/data/tenet_cv2.pkl', 'wb') as file:
    pickle.dump(cv2, file)
```

```python
cv2.get_feature_names()         # 단어목록
```

```python
cv2.vocabulary_
```

```python
# 나이브 베이즈 이론

nb = MultinomialNB()
nb.fit(x_train_cv, tenet_train.y)
```

```python
# test
tenet_test_x_cv = cv2.transform(tenet_test.x)
y_predict2 = nb.predict(tenet_test_x_cv)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(tenet_test.y, y_predict2)
```

```python
print(classification_report(tenet_test.y, y_predict2))
```

```python
pd.crosstab(tenet_test.y, y_predict2)
```
