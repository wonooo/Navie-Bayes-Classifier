```python
# 샘플링은 할때마다 데이터가 바뀌기 때문에 한번 돌린 후 저장한 파일을 이용해 혼란을 줄임
import pickle
# 바이너리파일 불러오기
with open('c:/data/tenet_x_train.pkl', 'rb') as file:
    x_train = pickle.load(file)
# 바이너리파일 불러오기
with open('c:/data/tenet_x_test.pkl', 'rb') as file:
    x_test = pickle.load(file)
# 바이너리파일 불러오기
with open('c:/data/tenet_y_train.pkl', 'rb') as file:
    y_train = pickle.load(file)
# 바이너리파일 불러오기
with open('c:/data/tenet_y_test.pkl', 'rb') as file:
    y_test = pickle.load(file)
```

```python
# 나이브베이즈 이론을 학습시키기

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
cv = CountVectorizer(ngram_range=(1,1))
x_train_cv = cv.fit_transform(x_train)   
```

```python
x_train_cv
```

```python
# 학습시킨 데이터 바이너리파일로 저장
import pickle
with open('c:/data/tenet_cv.pkl', 'wb') as file:
    pickle.dump(cv, file)
```

```python
# 단어목록 확인
cv.get_feature_names()         
```

```python
# 각각 단어들의 빈도 확인
cv.vocabulary_
```

```python
# 나이브베이즈 이론에 적용

nb = MultinomialNB()
nb.fit(x_train_cv, y_train)
```

```python
# 다음파일에 대해서 적용하기 위해 바이너리파일 저장

with open('c:/data/tenet_nb.pkl', 'wb') as file:
    pickle.dump(nb, file)
```

```python
# test
x_test_cv = cv.transform(x_test)

x_test_cv
```

```python
y_predict = nb.predict(x_test_cv)
```

```python
y_predict == y_test
```

```python
accuracy_score(y_test, y_predict)
```

```python
import pandas as pd
pd.crosstab(y_test, y_predict)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_predict))
```
